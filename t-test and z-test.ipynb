```python
from scipy import stats

# Example data for two groups
group1 = [25, 30, 35, 40, 45]
group2 = [20, 28, 32, 38, 42]

# Perform two-sample t-test
statistic, p_value = stats.ttest_ind(group1, group2)

# Display the results
print("T-statistic:", statistic)
print("P-value:", p_value)

# Check the significance level (e.g., alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Alternative hypothesis: There is a significant difference between the groups.")
else:
    print("null hypothesis: There is no significant difference between the groups.")
```

```python
import statsmodels.api as sm
import numpy as np

# Example data for two groups
group1 = [25, 30, 35, 40, 45]
group2 = [20, 28, 32, 38, 42]

# Perform two-sample Z-test
statistic, p_value = sm.stats.ztest(np.array(group1), np.array(group2))

# Display the results
print("Z-statistic:", statistic)
print("P-value:", p_value)

# Check the significance level (e.g., alpha = 0.05)
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis: There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the groups.")
```

```python
import statsmodels.api as sm
import numpy as np

# Example data for two groups
data = pd.read_csv('movies_matplotlib_dataset.csv')
data= data.dropna()
# Perform two-sample Z-test
statistic, p_value = sm.stats.ztest(data['budget'], data['gross'])

# Display the results
print("Z-statistic:", statistic)
print("P-value:", p_value)

# Check the significance level (e.g., alpha = 0.05)
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis: There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the groups.")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

df = pd.read_csv('movies_matplotlib_dataset.csv')

# Contingency table
contingency_table = pd.crosstab(df['rating'], df['country'])

# Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print results
print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

# Interpret the results
if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant association between 'rating' and 'country'.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between 'rating' and 'country'.")
```
